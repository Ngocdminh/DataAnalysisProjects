```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\HR analytics\merged_data.csv'
data = pd.read_csv(file_path)
```

```python
## Create dataframe
df = pd.DataFrame(data)
```

```python
## Set the aesthetic style of the plots
sns.set_style('whitegrid')
```

```python
## Creaet a pairplot to visualize the relationships between employee's performance and personal attributes
##pairplot = sns.pairplot(df, hue = 'ManagerRating')
##plt.suptitle('Pairplot of employee performance')

##Create a boxplot to visualize the distribution of performance ratings by different categories
plt.figure(figsize=(15,20))

plt.subplot(2,2,1)
sns.boxplot(x='Age', y ='ManagerRating', data=df)
plt.title('Performance Rating by Age')

plt.subplot(2,2,2)
sns.boxplot(x='Gender', y ='ManagerRating', data=df)
plt.title('Performance Rating by Gender')

plt.subplot(2,2,3)
sns.boxplot(x='MaritalStatus', y='ManagerRating', data=df)
plt.title('Performance Rating by Marital Status')

plt.subplot(2,2,4)
sns.boxplot(x='Ethnicity', y='ManagerRating', data=df)
plt.title('Performance Rating by Ethnicity')

plt.tight_layout()
plt.show()
```
