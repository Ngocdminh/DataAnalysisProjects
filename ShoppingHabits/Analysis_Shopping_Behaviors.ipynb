```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/merged_data.csv')

##Analyze product sales by type of items and product category
product_sales = df.groupby(['item_purchased','Category'])['purchase_amount'].sum().reset_index() ##group data by category and type of items, sum total purchase_amount, then reset dataframe
print(product_sales)

plt.figure(figsize=(20,12))
plt.bar(product_sales['item_purchased'], product_sales['purchase_amount'])
plt.title('Sales by product category')
plt.xlabel('Product Category')
plt.ylabel('Sales amount (in units)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
