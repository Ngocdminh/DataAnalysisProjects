```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\SwedishHousingPrices\SwedenHousingPrices.csv'
data = pd.read_csv(file_path)
```

```python
## Convert date_published to datetime format
data['date_published'] = pd.to_datetime(data['date_published'])
## Set date_published as the index
data.set_index('date_published', inplace=True)
```

```python
## Resample the data to monthly frequency and calculate the mean asking price
monthly_prices = data['asking_price_sek'].resample('ME').mean()

## Calculate the rolling mean (trend)
##rolling_mean = monthly_prices.rolling(window=12).mean()
```

```python
## Plot the time series of monthly average asking prices
plt.figure(figsize=(12,6))
plt.plot(monthly_prices, marker='o', linestyle='-')
##plt.plot(rolling_mean, color='red', linestyle='-', label='Trend (12-month Rolling Mean)')
plt.title('Monthly Average Asking Prices')
plt.xlabel('Date')
plt.ylabel('Average Asking Price (SEK)')
###plt.legend()
plt.grid(True)
plt.show()
```
