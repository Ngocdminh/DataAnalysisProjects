```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\SwedishHousingPrices\SwedenHousingPrices.csv'
data = pd.read_csv(file_path)
```

TYPOLOGY ANALYSIS: compare average prices, land area, and living area across different typology categories

```python
## Group by typology and calculate mean values for relevant columns
typology_analysis = data.groupby('typology').agg({'asking_price_sek':'mean','land_area_sqm':'mean','living_area_sqm':'mean'}).reset_index()

## Display the results
print(typology_analysis)
```

```python
## Sort the data by asking price
typology_analysis = typology_analysis.sort_values(by='asking_price_sek')
## Plot the average asking price by typology
plt.figure(figsize=(10,6))
sns.barplot(x='typology', y='asking_price_sek', data=typology_analysis)
plt.title('Average Land Area by Typology')
plt.xlabel('Typlogy')
plt.ylabel('Average Land Area (sqm)')
plt.xticks(rotation=45)
plt.show()
```

ROOM COUNT ANALYSIS: how the number of rooms affects the asking price

```python
#Group by 'number_rooms' and calculate mean asking price
room_count_analysis=data.groupby('number_rooms').agg({'asking_price_sek':'mean'}).reset_index()
print(room_count_analysis)

## Sort the data by asking price
room_count_analysis = room_count_analysis.sort_values(by='asking_price_sek')

## Plot the average asking price by number of rooms
plt.figure(figsize=(14,6))
sns.barplot(x='number_rooms',y='asking_price_sek', data=room_count_analysis, order=room_count_analysis['number_rooms'])
plt.title('Average Asking Price by Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Average Asking Price (SEK)')
plt.show()
```
