```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\SwedishHousingPrices\SwedenHousingPrices.csv'
data = pd.read_csv(file_path)
```

```python
## Convert 'coordenates' to separate latitude and longitude columns
data[['latitude','longitude']] = data['coordenates'].str.split(',',expand=True).astype(float)
```

```python
## Plot the scatter plot of housing prices by location
plt.figure(figsize=(10,10))
plt.scatter(data['longitude'], data['latitude'], c=data['asking_price_sek'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Asking Price (SEK)')
plt.title('Housing Prices by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
