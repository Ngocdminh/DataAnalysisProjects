```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\SwedishHousingPrices\SwedenHousingPrices.csv'
data = pd.read_csv(file_path)
```

```python
## Calculate summary statistics for nmerical columns like mean, median, standard deviations
summary_stats=data.describe()
print(summary_stats)
```

```python
## Plot histogram for numerical columns to visualize their distributions
numerical_columns=['asking_price_sek', 'land_area_sqm', 'living_area_sqm', 'sqm_price_sek']
data[numerical_columns].hist(bins=30, figsize=(15,10))
plt.suptitle('Histogram of Numerical Columns')
plt.show()

##Plot box plots for numerical columns to identify the spread and potential outliers in numerical forms
plt.figure(figsize=(15,10))
##Create a 2x2 grid of box plots for each numerical column in the dataset, to visualize the spread, central tendency, and potential outliers 
for i, column in enumerate(numerical_columns,1):
    plt.subplot(2,2,i)   ##creates a subplot in a 2x2 grid
    sns.boxplot(y=data[column]) ##creates a box plot for the current column, specifies that the data comes from the current column of the DataFrame.
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
