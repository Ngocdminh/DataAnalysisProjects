```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Load the csv file
file_path = r'C:\Users\ngocd\source\Github\DataAnalysisProjects\SwedishHousingPrices\SwedenHousingPrices.csv'
data = pd.read_csv(file_path)
```

```python
## Select numerical features for correlation analysis
numerical_features = ['asking_price_sek', 'land_area_sqm', 'living_area_sqm', 'sqm_price_sek', 'number_rooms']

## Calculate the correlation matrix
correlation_matrix = data[numerical_features].corr()

## Display the correlation matrix
print(correlation_matrix)
```

```python
## Plot the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f') ##annot=True adds correlation coefficients to heatmap, cmap sets the color scheme
plt.title('Correlation Heatmap')
plt.show()
```
